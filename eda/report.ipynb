## max i min dla poszczegolnego parametru time acc, pre, recall, f1 roc, z podziałem na model scaler i balancing

```python
score_columns = ['time_training', 'accuracy_score', 'precision_score', 'recall_score', 'f1_score', 'roc_auc_score']

for col in score_columns:
    print(f"--- Analysis for: {col} ---")

    max_row = data_non_params.loc[data_non_params[col].idxmax()]
    print(f"Max {col}:")
    print(f"Model: {max_row.get('model')}, Scaler: {max_row.get('scaler')}, Balancing name: {max_row.get('balancing name')}, Value: {max_row[col]}")

    print("-" * 20)

    min_row = data_non_params.loc[data_non_params[col].idxmin()]
    print(f"Min {col}:")
    print(f"Model: {min_row.get('model')}, Scaler: {min_row.get('scaler')}, Balancing name: {min_row.get('balancing name')}, Value: {min_row[col]}")

    print("\n")
```
